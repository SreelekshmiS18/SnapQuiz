```python
import tkinter as tk
from tkinter import messagebox

questions = [
    {"question": "What is the capital of France?", "choices": ["Berlin", "Madrid", "Paris", "Rome"], "answer": "Paris"},
    {"question": "Which planet is known as the Red Planet?", "choices": ["Earth", "Mars", "Jupiter", "Venus"], "answer": "Mars"},
    {"question": "Who wrote Romeo and Juliet?", "choices": ["Charles Dickens", "William Shakespeare", "Mark Twain", "Jane Austen"], "answer": "William Shakespeare"},
    {"question": "What is the largest mammal in the world?", "choices": ["Elephant", "Blue Whale", "Giraffe", "Great White Shark"], "answer": "Blue Whale"},
    {"question": "Which element does the chemical symbol 'O' represent?", "choices": ["Gold", "Oxygen", "Silver", "Iron"], "answer": "Oxygen"},
    {"question": "What is the hardest natural substance on Earth?", "choices": ["Gold", "Iron", "Diamond", "Platinum"], "answer": "Diamond"},
    {"question": "Which country is known as the Land of the Rising Sun?", "choices": ["China", "Japan", "South Korea", "Thailand"], "answer": "Japan"},
    {"question": "What is the smallest prime number?", "choices": ["1", "2", "3", "5"], "answer": "2"},
    {"question": "Who painted the Mona Lisa?", "choices": ["Vincent van Gogh", "Pablo Picasso", "Leonardo da Vinci", "Claude Monet"], "answer": "Leonardo da Vinci"},
    {"question": "What is the boiling point of water at sea level?", "choices": ["90°C", "100°C", "110°C", "120°C"], "answer": "100°C"}
]

current_qn_index = 0
score = 0

def check_answer():
    global current_qn_index, score
    if selected.get() == questions[current_qn_index]["answer"]:
        score += 1
    current_qn_index += 1
    if current_qn_index < len(questions):
        load_next_question()
    else:
        messagebox.showinfo("Quiz Finished", f"Your score is: {score}/{len(questions)}")
        window.destroy()

def load_next_question():
    lbl_question.config(text=questions[current_qn_index]["question"])
    selected.set(None)
    for rb, choice in zip(option_buttons, questions[current_qn_index]["choices"]):
        rb.config(text=choice, value=choice)

window = tk.Tk()
window.title("Quiz Maker")

lbl_title = tk.Label(window, text="SnapQuiz", font=("Arial Bold", 24))
lbl_title.pack()


lbl_question = tk.Label(window, text=questions[current_qn_index]["question"], font=("Arial Bold", 15))
lbl_question.pack()


selected = tk.StringVar()
option_buttons = []
for choice in questions[current_qn_index]["choices"]:
    rb = tk.Radiobutton(window, text=choice, value=choice, variable=selected)
    rb.pack(anchor="w")
    option_buttons.append(rb)


submit_button = tk.Button(window, text="Submit", command=check_answer,bg="grey")
submit_button.pack()

window.mainloop()
```
